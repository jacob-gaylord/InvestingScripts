```python
import websocket
import json
import pandas as pd
```

```python
endpoint = 'wss://testnet.binancefuture.com/ws'
```

```python
our_msg = json.dumps({
"method": "SUBSCRIBE",
"params":
[
"btcusdt@kline_5m",
"btcusdt@kline_15m"
],
"id": 1
})
```

```python
returns = {'5m':0, '15m':0}
```

```python
in_position = False
```

```python
def on_open(ws):
    ws.send(our_msg)
    
def on_message(ws, message):
    global df_, in_position, buyprice
    out = json.loads(message)
    df_ =pd.DataFrame(out['k'],index=[pd.to_datetime(out['E'],
                                            unit='ms')])[['s', 'i', 'o', 'c']]
    df_['ret '+df_.i.values[0]] = float(df_.c)/float(df_.o) -1
    returns[df_.i.values[0]] = float(df_.c)/float(df_.o) -1
    print(df_)
    if not in_position and returns['15m'] < 0 and returns['5m'] > 0:
        buyprice = float(df_.c)
        in_position = True
        print('bought @' + str(buyprice))
    if in_position:
        print('in position, check selling')
        print('target profit: ' + str(buyprice * 1.0002))
        print('stop loss: ' + str(buyprice * 0.9998))
        if float(df_.c) > buyprice * 1.0002:
            print('target profit reached - sell')
            print('profit: ' + str(float(df_.c) - buyprice))
            in_position = False
        elif float(df_.c) < buyprice * 0.9998:
            print('stop loss reached - sell')
            print('loss: ' + str(float(df_.c) - buyprice))
            in_position = False
    
ws = websocket.WebSocketApp(endpoint, on_message=on_message,
                           on_open=on_open)
ws.run_forever()
```
