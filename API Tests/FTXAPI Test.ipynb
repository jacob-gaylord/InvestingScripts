```python
import requests
import websocket
import json
import pandas as pd
```

```python
endpoint = 'wss://ftx.com/ws/'
```

```python
our_msg = json.dumps({'op': 'subscribe', 'channel': 'trades', 'market': 'BTC/USD'})
```

```python
def on_open(ws):
    ws.send(our_msg)
    
def on_message(ws, message):
    out = json.loads(message)
    df_ = pd.DataFrame(out['data'], index=[0])
    df_.index = pd.to_datetime(df_.time)
    print(df_['price'])
    
ws = websocket.WebSocketApp(endpoint, on_message=on_message,
                           on_open=on_open)
ws.run_forever()
```
